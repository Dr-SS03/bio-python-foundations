```python
import pandas as pd
import numpy as np

# Simulated gene expression counts
df = pd.DataFrame({
    "gene": ["TP53", "BRCA1", "MYC", "EGFR"],
    "sample_1": [120, 10, 300, 50],
    "sample_2": [140, 12, 280, 45]
})

# Compute mean expression
df["mean_counts"] = df[["sample_1", "sample_2"]].mean(axis=1)

# Log transform (common in RNA-seq)
df["log2_mean"] = np.log2(df["mean_counts"] + 1)

df.sort_values("mean_counts", ascending=False)
```
